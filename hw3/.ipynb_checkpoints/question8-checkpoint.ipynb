```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import io
```

```python
complete_dataset = {}
for data_name in ["mnist"]:
    data = io.loadmat("data/%s_data.mat" % data_name)
    print("\nloaded %s data!" % data_name)
    complete_dataset[data_name] = {}
    
    fields = "test_data", "training_data", "training_labels"
    for field in fields:
        print(field, data[field].shape)
        complete_dataset[data_name][field] = data[field]
```

# Part 1

### 1.1 Loading Data and Contrast-Normalizing the images

```python
mnist_train_x_denorm = complete_dataset["mnist"]["training_data"]
mnist_train_y = complete_dataset["mnist"]["training_labels"]

## Contrast normalizing data:
mnist_train_x = mnist_train_x_denorm / np.linalg.norm(mnist_train_x_denorm, axis=1)[:, None]

mnist_train_x.shape
mnist_train_y.shape
```

### 1.2 Computing mean and covariance for each class

```python
mean_samples = {}
cov_samples = {}

for digit in list(np.unique(mnist_train_y)):
    data_for_digit = mnist_train_x[mnist_train_y.reshape(-1,) == digit, :]
    
    mean = data_for_digit.mean(axis=0)
    covariance = np.cov(data_for_digit.T) 
    
    mean_samples[digit] = mean
    cov_samples[digit] = covariance
```

# Part 2

```python
fig= plt.figure(figsize=(7,7))
plt.imshow(cov_samples[8])
plt.colorbar()
plt.show()
```

It is clear from the plot that the covariances on the off-diagonal are less than the covariances on the diagonal. We can conclude that the same pixel from each image (of class D) behave in a similar fashion. Thus, they will be quite close to eachother in terms of magnitude. Then, the features that correspond to the mainly-negative off-diagonal covariances behave in an opposite manner. In terms of the image of class D, if cov(X1,X2) < 0, then while feature X1 is "on"/white, feature X2 will usually be "off"/black (or vice versa). 

# Part 3

Partioning

```python
def partitionDataSet(training_data, training_labels, length): 
    totalSet = np.concatenate((training_data, training_labels), axis=1) 
    np.random.shuffle(totalSet)
    return totalSet[0:length, :], totalSet[length:, :]
```

```python
# Partioning dataset
mnist_train_x = complete_dataset["mnist"]["training_data"]
mnist_train_y = complete_dataset["mnist"]["training_labels"]

mnist_val, mnist_train = partitionDataSet(mnist_train_x, mnist_train_y, 10000)

mnist_val_x = mnist_val[:, :-1]
mnist_val_y = mnist_val[:, -1:]

mnist_train_x = mnist_train[:, :-1]
mnist_train_y = mnist_train[:, -1:]

## Contrast Normalization
norm_train = np.sqrt((mnist_train_x**2).sum(axis=1))
mnist_train_x = mnist_train_x / norm_train.reshape(-1, 1)

norm_val = np.sqrt((mnist_val_x**2).sum(axis=1))
mnist_val_x = mnist_val_x / norm_val.reshape(-1, 1)

mnist_test_x = complete_dataset["mnist"]["test_data"]
norm_test = np.sqrt((mnist_test_x**2).sum(axis=1))
mnist_test_x = mnist_test_x / norm_test.reshape(-1, 1)
```

## 3a)  LDA

```python

training_ints = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
accuracies = {"train": [], "val": []}
predictions = []

## Computes prior, mean, and covariance
def train_lda(train_x, train_y):
#     print(train_x.shape)
#     print(train_y.shape)
    n,d = train_x.shape
    priors = np.zeros((10,1))
    covariance = np.zeros((d,d)).astype(np.float32)
    mean_samples = np.zeros((10, d))
    train_y = train_y.reshape(-1,)
    
    for digit in range(0, 10):
        ## Compute priors
        prob = (train_y == digit).astype(np.int32).sum() / n
        priors[int(digit)] = prob

        data = train_x[train_y == digit, :]
         
        mean_samples[int(digit),:] = data.mean(axis=0)
        covariance += np.cov(data.T)
        
    covariance /= 10
    return priors, mean_samples, covariance

def normalize_data(data, mean, covariance):
    mean_center = data - mean.T
    covariance_norm = np.sqrt((covariance**2).sum(axis=1)).reshape(1, -1)
    return mean_center / covariance_norm

# Train
for num_training in training_ints:
    print(f"Training with {num_training} examples") 

    train_x = mnist_train_x[:num_training, :]
    train_y = mnist_train_y[:num_training, :]
    
    priors, means, covariance = train_lda(train_x, train_y)
    
    u, sig, v = np.linalg.svd(covariance)
    covariance = covariance + (sig.min())*np.eye(*covariance.shape)
    
    N_train = mnist_train_x.shape[0]
    out_train = np.zeros((N_train, 10))
    
    N_val = mnist_val_x.shape[0]
    out_val = np.zeros((N_val, 10))
        
    for class_y in range(0, 10):
        prior = priors[class_y]
        mean = means[class_y, :].reshape(-1, 1)
        
        w = np.linalg.solve(covariance.T, mean)
        alpha = -0.5 * w.T.dot(mean) + np.log(prior)

        out_train[:, class_y] = (mnist_train_x.dot(w) + alpha).reshape(-1,)
        out_val[:, class_y] = (mnist_val_x.dot(w) + alpha).reshape(-1,)
    
    train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
    val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)
    
#     print("train",(mnist_train_y == train_pred).mean())
#     print("val",(mnist_val_y == val_pred).mean())
    predictions.append(val_pred)
    accuracies["train"].append((mnist_train_y == train_pred).mean())
    accuracies["val"].append((mnist_val_y == val_pred).mean())
    
    print(f"Training acc: {accuracies['train'][-1]}; Validation acc: {accuracies['val'][-1]}\n")
    print()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 7))
axes.plot(training_ints, [1 - x for x in accuracies["train"]], '.r-')
axes.plot(training_ints, [1 - x for x in accuracies["val"]], '.b-')
axes.legend(['Training', 'Validation'], loc='upper right')
axes.set_title("Error Rate vs. # Training Samples used")
axes.set_xlabel("Training Samples Used")
axes.set_ylabel("Error Rate")
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 7))

error_rate = {}
for pred in predictions:
    for digit in range(0, 10):
        digit_predictions = pred[mnist_val_y == digit]
        labels = mnist_val_y[mnist_val_y == digit]
        error_rate[digit] = error_rate.get(digit, []) + [(digit_predictions == labels).mean()]

        
for digit in error_rate.keys():
    axes.plot(training_ints, [1 - x for x in error_rate[digit]])
    
axes.legend([f"Digit {x}" for x in list(range(0, 10))], loc='upper right')
axes.set_title("Digit-wise Error Rate vs. # Training points used")
axes.set_xlabel("Training points Used")
axes.set_ylabel("Error Rate")
axes.set_xticklabels(training_ints)
plt.show()
```

### 3b) QDA

```python
training_ints = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
accuracies = {"train": [], "val": []}
predictions = []

## Computes prior, mean, and covariance
def train_qda(train_x, train_y):
    n,d = train_x.shape
    priors = np.zeros((10,1))
    covariances = np.zeros((10,d,d)).astype(np.float32)
    mean_samples = np.zeros((10, d))
    
    digits = list(np.unique(train_y))
    
    for digit in digits:
        prob = (train_y == digit).astype(np.int32).sum() / n
        priors[int(digit)] = prob

        data = train_x[train_y == digit, :]
        
        mean_samples[int(digit),:] = data.mean(axis=0)
        covariances[int(digit), :, :] = np.cov(data.T)
        
    return priors, mean_samples, covariances

def computeDiagonals(X, X_T):
    N,D = X.shape
    out_diag = np.zeros((N,1))
    
    for n in range(N):
        i = X[n, :].reshape(1, -1)
        i_t = X_T[:, n].reshape(-1, 1)
        out_diag[n] = i.dot(i_t)
    
    return out_diag
# Train
for num_training in training_ints:
    print(f"Training with {num_training} examples") 

    train_x = mnist_train_x[:num_training, :]
    train_y = mnist_train_y[:num_training, :].reshape(-1,)  
    
    priors, means, covariances = train_qda(train_x, train_y)
    
    N_train = mnist_train_x.shape[0]
    out_train = np.zeros((N_train, 10))
    
    N_val = mnist_val_x.shape[0]
    out_val = np.zeros((N_val, 10))
        
    for digit in range(0, 10):
        prior = priors[digit]
        mean = means[digit, :]
        covariance = covariances[digit, :, :]
        
        ## Adding a perturbation to make covariance singular
        u, sig, v = np.linalg.svd(covariance)
        covariance = covariance + (sig.min()*(10**-1)) * np.eye(*covariance.shape)
        alpha = -0.5 * np.linalg.det(covariance) + prior
        
        mean_centered_train = mnist_train_x - mean
        weight_train = np.linalg.solve(covariance, mean_centered_train.T)
        prediction_train = -0.5*computeDiagonals(mean_centered_train, weight_train) + alpha
        out_train[:, digit] = prediction_train.reshape(N_train,)
        
        mean_centered_val = mnist_val_x - mean
        weight_val = np.linalg.solve(covariance, mean_centered_val.T)
        prediction_val = -0.5*computeDiagonals(mean_centered_val, weight_val) + alpha
        out_val[:, digit] = prediction_val.reshape(N_val,)
        
        print(f"{digit}", end='-')
                
    print()    
    train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
    val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)
    
    predictions.append(val_pred)
    accuracies["train"].append((mnist_train_y == train_pred).mean())
    accuracies["val"].append((mnist_val_y == val_pred).mean())
    
    print(f"Training acc: {accuracies['train'][-1]}; Validation acc: {accuracies['val'][-1]}\n")
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 7))
axes.plot(training_ints, [1 - x for x in accuracies["train"]], '.r-')
axes.plot(training_ints, [1 - x for x in accuracies["val"]], '.b-')
axes.legend(['Training', 'Validation'], loc='upper right')
axes.set_title("Error Rate vs. # Training Samples used")
axes.set_xlabel("Training Samples Used")
axes.set_ylabel("Error Rate")
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(7, 7))

error_rate = {}
for pred in predictions:
    for digit in range(0, 10):
        digit_predictions = pred[mnist_val_y == digit]
        labels = mnist_val_y[mnist_val_y == digit]
        error_rate[digit] = error_rate.get(digit, []) + [(digit_predictions == labels).mean()]

        
for digit in error_rate.keys():
    axes.plot(training_ints, [1 - x for x in error_rate[digit]])
    
axes.legend([f"Digit {x}" for x in list(range(0, 10))], loc='upper right')
axes.set_title("Digit-wise Error Rate vs. # Training points used")
axes.set_xlabel("Training points Used")
axes.set_ylabel("Error Rate")
plt.show()
```

# 4) Spam Classifier

```python
spam_dataset = {}

data = io.loadmat("data/%s_data.mat" % "spam")
print("\nloaded %s data!" % data_name)

fields = "test_data", "training_data", "training_labels"
for field in fields:
    print(field, data[field].shape)
    spam_dataset[field] = data[field]
```

```python
spam_train_x = spam_dataset["training_data"]
spam_train_y = spam_dataset["training_labels"]

spam_val, spam_train = partitionDataSet(spam_train_x, spam_train_y, 500)

spam_val_x = spam_val[:, :-1]
spam_val_y = spam_val[:, -1:]

spam_train_x = spam_train[:, :-1]
spam_train_y = spam_train[:, -1:]

spam_test_x = spam_dataset["test_data"]
```

## 4 a) LDA on Spam Dataset

```python
## Computes prior, mean, and covariance
def spam_train_lda(train_x, train_y):
    n,d = train_x.shape
    priors = np.zeros((2,1))
    covariance = np.zeros((d,d)).astype(np.float32)
    mean_samples = np.zeros((2, d))
    train_y = train_y.reshape(-1,)
    
    for spam in [0,1]:
        ## Compute priors
        prob = (train_y == spam).astype(np.int32).sum() / n
        priors[int(spam)] = prob

        data = train_x[train_y == spam, :]        
        mean_samples[int(spam),:] = data.mean(axis=0)
        covariance += np.cov(data.T)
        
    covariance /= 2
    return priors, mean_samples, covariance

# Train
lda_priors, lda_means, lda_covariance = spam_train_lda(spam_train_x, spam_train_y)

N_train = spam_train_x.shape[0]
out_train = np.zeros((N_train, 2))

N_val = spam_val_x.shape[0]
out_val = np.zeros((N_val, 2))
        
for class_y in [0,1]:
    prior = lda_priors[class_y]
    mean = lda_means[class_y, :].reshape(-1, 1)

    w = np.linalg.solve(lda_covariance.T, mean)
    alpha = -0.5 * w.T.dot(mean) + np.log(prior)

    out_train[:, class_y] = (spam_train_x.dot(w) + alpha).reshape(-1,)
    out_val[:, class_y] = (spam_val_x.dot(w) + alpha).reshape(-1,)

train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)

print(f"Training acc: {(spam_train_y == train_pred).mean()}; Validation acc: {(spam_val_y == val_pred).mean()}\n")
```

## 4 b) QDA On Spam

```python
## Computes prior, mean, and covariance
def spam_train_qda(train_x, train_y):
    n,d = train_x.shape
    priors = np.zeros((2,1))
    covariances = np.zeros((2,d,d)).astype(np.float32)
    mean_samples = np.zeros((2, d))
    train_y = train_y.reshape(-1,)
    
    for spam in [0,1]:
        ## Compute priors
        prob = (train_y == spam).astype(np.int32).sum() / n
        priors[int(spam)] = prob

        data = train_x[train_y == spam, :] 
        
        mean_samples[int(spam),:] = data.mean(axis=0)
        covariances[int(spam), :, :] = np.cov(data.T)
        
    return priors, mean_samples, covariances

# Train
qda_priors, qda_means, qda_covariances = spam_train_qda(spam_train_x, spam_train_y)

N_train = spam_train_x.shape[0]
out_train = np.zeros((N_train, 2))

N_val = spam_val_x.shape[0]
out_val = np.zeros((N_val, 2))

for spam in [0,1]:
    prior = qda_priors[spam]
    mean = qda_means[spam, :]
    covariance = qda_covariances[spam, :, :]

    u, sig, v = np.linalg.svd(covariance)
    covariance = covariance + (sig.min()*(10**-1)) * np.eye(*covariance.shape)
    
    alpha = -0.5 * np.linalg.det(covariance) + prior

    mean_centered_train = spam_train_x - mean
    weight_train = np.linalg.solve(covariance, mean_centered_train.T)
    prediction_train = -0.5*computeDiagonals(mean_centered_train, weight_train) + alpha
    out_train[:, spam] = prediction_train.reshape(N_train,)

    mean_centered_val = spam_val_x - mean
    weight_val = np.linalg.solve(covariance, mean_centered_val.T)
    prediction_val = -0.5*computeDiagonals(mean_centered_val, weight_val) + alpha
    out_val[:, spam] = prediction_val.reshape(N_val,)


train_pred = np.argmax(out_train, axis=1).reshape(-1, 1)
val_pred = np.argmax(out_val, axis=1).reshape(-1, 1)

print(f"Training acc: {(spam_train_y == train_pred).mean()}; Validation acc: {(spam_val_y == val_pred).mean()}\n")
```
